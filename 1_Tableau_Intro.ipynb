#### Why Visualizations
3 Simple Reasons Why You Should First Visualize Data Before Doing Anything Else

- <b>Understand the Missingness Pattern</b> - Real-world data have the missingness problem for various reasons. Subjects refuse to answer specific questions due to privacy concerns like race and income. Subjects accidentally enter the incorrect data (e.g., missing a digit in the phone number).

- <b>Identify Outliers </b>- One early step in data pre-processing is to handle outliers, which are data values that are distinctly different from other data values in the dataset. They can happen at least for two reasons. First, the data are entered by mistake. For example, missing a decimal can make the value 100 times larger (10.24 vs. 1024).

- <b>Generate Meaningful Hypotheses</b> - As machine learning setting, it's always necessary to form
some hypotheses and develop some meaningful models with the training dataset, which we hope hold when we run the models in the test dataset. Similar to the identification of outliers, specific data structures of given datasets will determine how we visualize the data before doing any testing. For example, if you're trying to identify a possible linear relationship between two continuous variables, you can start with the scatter plots, as shown in the last section.

#### Data Pane
- dimensions : The columns on which we are not going to perform operations. we use this as reference are called dimensions
- measures : The columns in our dataset on which we are going to perform mathematical operations are called measures

Tableau creates some default dimensions and measures like
- Measure Names(Dimensions)
- Latitude(Measures)
- Longitude(Measures)
- Number of Records(Measures)
